```python
import numpy as np
import librosa as li
import matplotlib.pyplot as plt
from IPython.display import Audio
```

```python
filename = li.util.example_audio_file()
x,fs = li.load(filename)
x = x[:5*fs]
```

```python
Audio(x,rate=fs)
```

```python
S = li.core.cqt(x)
S.shape
```

```python
plt.figure(figsize=(10,10))
plt.imshow(abs(S),origin="lower",cmap="jet")
```
